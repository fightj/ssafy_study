### 19590 마법사의 사냥 D2 / soso
```python
n = int(input())
arr = [[int(a) for a in input().split()]for _ in range(n)]
k = int(input())

di = [-1, -1, 1, 1] #좌상 우상 우하 좌하
dj = [-1, 1, 1, -1]

max_result = 0

for i in range(1, n+1):
    for j in range(1, n+1):
        result = 0
        for skill in range(1, k+1):
            for a in range(4):
                ni = i + di[a]*skill
                nj = j + dj[a]*skill
                if 0 <= ni <n and 0 <= nj <n:
                    result += arr[ni][nj]
        if result > max_result:
            max_result = result

print(max_result)

```

### 19591 폭탄 터뜨리기 D3 / 좋은 문제
```python
n, m = map(int, input().split())
k = int(input())
arr = [[a for a in input()]for _ in range(n)]

di = [0, 1, 0, -1] #우하좌상
dj = [1, 0, -1, 0]

bomb_idx =[]
for i in range(n):
    for j in range(m):
        if arr[i][j] == '@':
            bomb_idx.append([i, j])

for t in bomb_idx:
    for i, j in [t]:
        arr[i][j] = '%'
        not_go = []
        for power in range(1, k+1):
            for a in range(4):
                if a in not_go:
                    continue
                else:
                    ni = i + di[a] * power
                    nj = j + dj[a] * power
                    if 0 <= ni < n and 0 <= nj < m:

                        if arr[ni][nj] != '#':
                            arr[ni][nj] = '%'
                        if arr[ni][nj] == '#':
                            not_go.append(a)

for i in range(n):
    print(*arr[i], sep='')
```

### 18539 사각형 그리기 게임 D3 / 좋은 문제
```python
t = int(input())
for cnt in range(1, t+1):
    n = int(input())
    arr = [[int(a)for a in input().split()] for _ in range(n)]
    result_arr=[]
    for i in range(n):
        for j in range(n):
            start = arr[i][j]
            for k in range(i+1, n):
                for q in range(j,n):
                    if arr[k][q] == start:
                        result_arr.append((k-i+1)*(q-j+1))
    result = result_arr.count(max(result_arr))
    print(f'#{cnt} {result}')
```

### 2805 농작물 수확하기 D3 / 좋은 문제
```python
T = int(input())

for tc in range(1, T + 1):
    n = int(input())
    input_arr = [[a for a in input()]for _ in range(n)]

    arr=[]
    for line in input_arr:
        row = [int(a) for a in line]
        arr.append(row)
    
    result = 0
    if n ==1:
        result += arr[0][0]
    else:
        j = 0
        for i in range(n):
            if i < n//2:
                result += sum(arr[i][n//2-j:n//2+j+1])
                j += 1
            elif i == n//2:
                result += sum(arr[i][n // 2 - j:n // 2 + j + 1])
                j -= 1
            else:
                result += sum(arr[i][n//2-j:n//2+j+1])
                j -= 1

    print(f'#{tc} {result}')

```


### 19774. 회문(palindrome)
- row_arr만드는 법 숙지
```python
t = int(input())
for tc in range(1, t + 1):
    N, M = map(int, input().split())
    arr = [[a for a in input()]for _ in range(N)]
    row_arr = [''.join(arr[row][col] for row in range(N)) for col in range(N)]

    for string in arr:
        for i in range(N-M+1):
            arr1 = string[i:i+M]
            if arr1 == arr1[::-1]:
                print(f'#{tc}', end=' ')
                print(*arr1, sep='')
                break

    for string in row_arr:
        for i in range(N-M+1):
            arr1 = string[i:i+M]
            if arr1 == arr1[::-1]:
                print(f'#{tc}', end=' ')
                print(*arr1, sep='')
                break
```

### 19817 자동 수학 계산기(mid)
```python
data = input()
data_num_lst = []
data_lst = []
data_num = ''
for i in range(len(data)):
    if i == 0 and not data[i].isdigit():
        data_num += data[i]
    elif data[i].isdigit():
        data_num += data[i]
        if i == len(data)-1:
            data_num_lst.append(data_num)
    else:
        data_num_lst.append(data_num)
        data_lst.append(data[i])
        data_num = ''

result = int(data_num_lst[0])
for i in range(len(data_lst)):
    op = data_lst[i]
    num = int(data_num_lst[i + 1])
    if op == '+':
        result += num
    else:
        result -= num
print(result)
```

### 19783 괄호 친구들(mid) D3

```python
def cal(data, current_num):
    big_num = ''
    small_num = ''
    i = 0
    while i < len(data):
        if data[i] == '[':
            i += 1
            while i < len(data) and data[i] != ']':
                big_num += data[i]
                i += 1
            current_num += int(big_num)
            big_num = ''
        elif data[i] == '{':
            i += 1
            while i < len(data) and data[i] != '}':
                small_num += data[i]
                i += 1
            current_num *= int(small_num)
            small_num = ''
        else:
            i += 1
    return current_num


data = input().strip()
result = cal(data, 0)
print(result)

```

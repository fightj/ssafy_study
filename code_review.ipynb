````python

def find_start_end(maze, N):
    start = None
    end = None
    for i in range(N):
        for j in range(N):
            if maze[i][j] == '2':
                start = (i, j)
            elif maze[i][j] == '3':
                end = (i, j)
    return start, end

def is_valid_move(maze, visited, x, y, N):
    return (0 <= x < N and 0 <= y < N and maze[x][y] in ('0', '3') and not visited[x][y])

def dfs(maze, visited, x, y, end_x, end_y, N):
    if (x, y) == (end_x, end_y):
        return True

    visited[x][y] = True

    directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # 상하좌우
    for dx, dy in directions:
        nx, ny = x + dx, y + dy
        if is_valid_move(maze, visited, nx, ny, N):
            if dfs(maze, visited, nx, ny, end_x, end_y, N):
                return True

    return False

def solve(maze, N):
    start, end = find_start_end(maze, N)
    if not start or not end:
        return 0

    start_x, start_y = start
    end_x, end_y = end

    visited = [[False] * N for _ in range(N)]
    return 1 if dfs(maze, visited, start_x, start_y, end_x, end_y, N) else 0

# 입력 처리
T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    maze = [input().strip() for _ in range(N)]

    result = solve(maze, N)
    print(f'#{tc} {result}')   
#1 1
#2 1
#3 0
````

## 20007 레드 마운틴(low)
- input data:3
0 1 0
0 1 1
0 0 0

- 미로찾기 게임과 매우 유사. DFS를 활용하여 길을 찾을 때 까지 진행.

````python
def mountain(arr, N):
    stack = [(0, 0)]
    while stack:
        y, x = stack.pop()
        arr[y][x] = -1 # 지나간 길 표시
        dir = [(0,1), (1,0), (-1,0), (0,-1)]
        for dy, dx in dir:
            ny, nx = y +dy, x +dx
            if 0<= ny < N and 0 <= nx < N:
                if (ny, nx) == (N-1, N-1):
                    return 1 # 도착점을 찾으면 return 1
                elif arr[ny][nx] == 0:
                    stack.append((ny, nx))
    return 0



N = int(input())
arr = [list(map(int, input().split()))for _ in range(N)]

result = mountain(arr, N)
print(result)
````

## 20004 다빈치 민코드(mid)
- input data:7 3
1 5 4 -2 6 7 -1

- 풀이방법: 입력받은 n개의 숫자들을 리스트로 받아와서, 그중 m개로 생성할 수 있는 집합들을 모두 만든 다음 해당 인자들의 곱이 낮으면 기존 result리스트를 clear()하고 해당 인자들을 for문을 통해 result리스트에 append

````python
import itertools

n, m = map(int, input().split())
arr = list(map(int, input().split()))
arr_m_list = list(itertools.combinations(arr, m))
result = []
min_num = float('inf')

for qqq in arr_m_list:
    a = 1
    for i in range(len(qqq)):
        a *= qqq[i]
    if a < min_num:
        min_num = a
        result.clear()
        for i in range(len(qqq)):
            result.append(qqq[i])

result.sort()
print(*result)
````

## 20003 10을 만들자(low)

- n을 입력받고, num = [1,2,3,4,5,6,7,8,9]로초기화 시킨 리스트를 ```make_10_lst = list(itertools.product(num, repeat=n))```로 중복을 허용해서 1 1 1 ~ 9 9 9 까지 모든 경우의 수를 튜플형태로 만들어서 반환.
